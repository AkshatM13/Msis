```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.count = 0

    def isempty(self):
        return self.head is None

    def InsertatHead(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node
        self.count += 1

    def traverse(self):
        curr = self.head
        while curr is not None:
            print(curr.data, end="-> ")
            curr = curr.next
        print()

    def InsertatTail(self, value):
        new_node = Node(value)
        if self.isempty():
            self.InsertatHead(value)
            return
        curr = self.head
        while curr.next is not None:
            curr = curr.next
        curr.next = new_node
        self.count += 1

    def display(self):
        curr = self.head
        while curr is not None:
            print(curr.data, end=" ")
            curr = curr.next
        print()

    def InsertatGivenData(self, after, value):
        new_node = Node(value)
        curr = self.head

        while curr is not None:
            if curr.data == after:
                break
            curr = curr.next

        if curr is not None:
            new_node.next = curr.next
            curr.next = new_node
            self.count += 1
        else:
            print('Item not found')

    def deleteHead(self):
        if self.isempty():
            print('Empty Linked List')
        else:
            self.head = self.head.next
            self.count -= 1

    def deleteTail(self):
        if self.isempty():
            print('Empty Linked List')
            return
        curr = self.head
        if curr.next is None:
            self.deleteHead()
            return
        while curr.next.next is not None:
            curr = curr.next
        curr.next = None
        self.count -= 1

    def remove(self, value):
        if self.head is None:
            print('Empty Linked List')
            return

        if self.head.data == value:
            self.deleteHead()
            return

        curr = self.head

        while curr.next is not None:
            if curr.next.data == value:
                break
            curr = curr.next

        if curr.next is None:
            print('Not Found')
        else:
            curr.next = curr.next.next
            self.count -= 1

    def search(self, item):
        curr = self.head
        pos = 0

        while curr is not None:
            if curr.data == item:
                return pos
            curr = curr.next
            pos += 1

        return 'Not Found'
    def AddCommonElements(self, list1, list2):
        set1 = set()
        curr1 = list1.head

        # Traverse the first list and add its elements to the set
        while curr1 is not None:
            set1.add(curr1.data)
            curr1 = curr1.next

        curr2 = list2.head

        # Traverse the second list and check for common elements
        while curr2 is not None:
            if curr2.data in set1:
                self.InsertatTail(curr2.data)
            curr2 = curr2.next

    def append(self, data):
        if not self.head:
            self.head = Node(data)
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = Node(data)

    def has_cycle(head):
        slow = head
        fast = head
    
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
    
        return False

# Example usage
ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)

# Creating a cycle for testing
ll.head.next.next.next.next.next = ll.head.next.next

print("Contains Cycle:", has_cycle(ll.head))

# Example usage:
list1 = LinkedList()
list1.InsertatTail(1)
list1.InsertatTail(2)
list1.InsertatTail(3)
list1.InsertatTail(4)

list2 = LinkedList()
list2.InsertatTail(3)
list2.InsertatTail(4)
list2.InsertatTail(5)
list2.InsertatTail(6)

common_list = LinkedList()
common_list.AddCommonElements(list1, list2)

print("Common elements in the two lists:")
common_list.traverse()

# Example usage:
l = LinkedList()
l.InsertatHead(10)
l.InsertatTail(20)
l.InsertatTail(30)

print("Traverse after Insertions:")
l.traverse()  

l.InsertatGivenData(20, 25)
print("Traverse after InsertatGivenData:")
l.traverse() 

l.deleteHead()
print("Traverse after deleteHead:")
l.traverse()  

l.deleteTail()
print("Traverse after deleteTail:")
l.traverse()  

print("Search for 25:")
print(l.search(25))

print("Search for 100:")
print(l.search(100)) 

l.remove(25)
print("Traverse after removing 25:")
l.traverse() 

    
```

```python
class Queue:
    def __init__(self):
        self.items = [None] * 2  # Initialize with a default size of 2
        self.front_index = 0
        self.rear_index = 0
        self.count = 0

    def is_empty(self):
        return self.count == 0

    def is_full(self):
        return self.count == len(self.items)

    def enqueue(self, item):
        if self.is_full():
            self.resize(2 * len(self.items))  # Double the size of the list
        self.items[self.rear_index] = item
        self.rear_index = (self.rear_index + 1) % len(self.items)
        self.count += 1

    def dequeue(self):
        if self.is_empty():
            raise IndexError("dequeue from empty queue")
        value = self.items[self.front_index]
        self.items[self.front_index] = None  # Optional: Clear the slot
        self.front_index = (self.front_index + 1) % len(self.items)
        self.count -= 1
        # Optionally, shrink the list to save memory
        if 0 < self.count < len(self.items) // 4:
            self.resize(len(self.items) // 2)
        return value

    def peek(self):
        if self.is_empty():
            raise IndexError("peek from empty queue")
        return self.items[self.front_index]

    def size(self):
        return self.count

    def resize(self, new_capacity):
        old_items = self.items
        self.items = [None] * new_capacity
        for i in range(self.count):
            self.items[i] = old_items[(self.front_index + i) % len(old_items)]
        self.front_index = 0
        self.rear_index = self.count

# Example usage:
queue = Queue()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print("Queue after enqueuing 1, 2, 3:")
while not queue.is_empty():
    print(queue.dequeue())
```

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            raise IndexError("Pop from empty stack")

    def is_empty(self):
        return len(self.stack) == 0

def reverse_file_content(filename):
    stack = Stack()

    # Read the file and push each line onto the stack
    with open(filename, 'r') as file:
        for line in file:
            stack.push(line.rstrip('\n'))

    # Write the lines back to the file in reverse order
    with open(filename, 'w') as file:
        while not stack.is_empty():
            file.write(stack.pop() + '\n')

# Example usage
filename = 'example.txt'
reverse_file_content(filename)
```
